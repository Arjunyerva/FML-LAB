```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
  return 1 / (1+np.exp(-x))
```

```python
x = np.linspace(-10, 10)
```

```python
plt.plot(x,  sigmoid(x))
plt.axis('tight')
```

```python
plt.title('Activation Function :sigmoid')
```

```python
plt.show()
```
